```python
import seaborn as sns
sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(5,5)})
```

```python
from scipy.stats import uniform
```

# Continuos Distribution 

```python
n=10000
start=0
width=20
data_uniform=uniform.rvs(size=n, loc=start,scale=width)
```

```python
ax=sns.distplot(data_uniform, bins=100,kde=True,color='red',hist_kws={"linewidth":15, 'alpha':1})
ax.set(xlabel='Uniform Distribution',ylabel='Frequency')
```

# Normal Distribution

```python
from scipy.stats import norm
#genrate random numbers from N(0,1)

data_normal=norm.rvs(size=10000,loc=0,scale=1)
```

```python
ax=sns.distplot(data_normal, bins=100,kde=True,color='blue',hist_kws={"linewidth":15, 'alpha':1})
ax.set(xlabel='Normal Distribution',ylabel='Frequency')
```

# Exponential Distribution

```python
from scipy.stats import expon
data_expon = expon.rvs(scale = 1, loc = 0, size = 1000)

ax=sns.distplot(data_expon, bins=100,kde=True,color='blue',hist_kws={"linewidth":15, 'alpha':1})
ax.set(xlabel = "Exponential Dist.", ylabel = "Frequency")
```

# Chi-Squared distribution


```python
from numpy import random
import matplotlib.pyplot as plt

x = random.chisquare(2, size = (2, 3))
print(x)

#sns.distplot(random.chisquare(df = 1, size = 1000), hist = False)
sns.distplot(random.chisquare(df=1,size=1000),hist=False)
plt.show()
```

# Weibull Distribution


```python
import numpy as np
a = 5
s = random.weibull(5, 10)


x = np.arange(1, 100) / 50

print(x)

def weib(x, n, a):
    return (a / n) * (x / n) ** (a - 1) * np.exp(- (x / n) ** a)

count, bins, ignored = plt.hist(np.random.weibull(5, 1000))
x = np.arange(1, 100) / 50
scale = count.max() / weib(x, 1, 5).max()
plt.plot(x, weib(x, 1, 5) * scale)
plt.show()
```
